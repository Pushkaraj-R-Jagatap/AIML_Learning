```python
import json
```

```python
 #load the data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
        
    return data
    
    
```

```python
data=load_data("store_data.json")
print(data)
print(type(data))
```

```python
#clean data
def clean_data(data):
    text_to_number={"one":1,"two":2,"three":3,"four":4,"five":5}
    for user in data:
       #clean ratings
        raw_rating=user["rating"].strip().lower()
        if(raw_rating in text_to_number):
            raw_rating=text_to_number[raw_rating]
        user["rating"]=raw_rating
        print(user["rating"])


        #handle missing value
        raw_age=user.get('age')
        if(raw_age==None):
            user["age"]=None
        
```

```python
clean_data(data)
```
