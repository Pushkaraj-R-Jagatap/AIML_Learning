```python
x=vendor_df[['Dollars']]
y=vendor_df['Freight']
```

```python
vendor_df.describe().round()
```

```python
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
model1=LinearRegression()
model1.fit(x_train,y_train)


model2=DecisionTreeRegressor(max_depth=3,random_state=42)
model2.fit(x_train,y_train)


model3=RandomForestRegressor(max_depth=3,random_state=42)
model3.fit(x_train,y_train)
```

```python
def evaluate_model(model,x_test,y_test,model_name):
    preds=model.predict(x_test)
    mae=mean_absolute_error(y_test,preds)
    rmse=mean_squared_error(y_test,preds)
    r2=r2_score(y_test,preds)*100



    print(f"\n{model_name} performance:")
    print(f"MAE     :{mae:.2f}")
    print(f"RMSE  :{rmse:.2f}")
    print(f"R2     :{r2:.2f}%")
```

```python
evaluate_model(model1,x_test,y_test,'linear_regression')
```

```python
evaluate_model(model2,x_test,y_test,'DecisionTreeRegressor')
```

```python
evaluate_model(model3,x_test,y_test,'RandomForestRegressor')
```
