```python
data=[]
for i in range (1,35):
    with open(f'scraped_data/review{i}.html','r',encoding='utf-8') as f:
        html_content=f.read()
        soup=BeautifulSoup(html_content,'html.parser')
        all_info = soup.select('div.fWi7J_')
       

        for info in all_info:
            # Get all text blocks with the common class inside this review
            items = info.select('div.css-146c3p1')
            
            # We must skip the elements that don't look like actual review cards 
            # (some .fWi7J_ elements are headers or filters)
            if len(items) < 5:
                continue
        
            # 1. Rating is usually the 1st block
            rating = items[0]
            
            # 2. Title is the 3rd block (skipping the separator dot at index 1)
            title = items[2]
            
            # 3. Review text is the 5th block (index 4)
            review = items[4]
            
            # 4. Name is found further down (index 5)
            name = items[5]
            
            # 5. City is index 6
            city = items[6]
            
            # 6. Date has a unique set of multiple classes
            Date = info.select_one('div.css-146c3p1.r-dnmrzs.r-1udh08x.r-1udbk01.r-3s2u2q.r-1iln25a')
        
            # Extracting Text
            Name = name.get_text(strip=True) if name else ""
            Title = title.get_text(strip=True) if title else ""
            Rating = rating.get_text(strip=True) if rating else ""
            City = city.get_text(strip=True).strip(', ') if city else ""
            Month = Date.get_text(strip=True).strip('· ') if Date else ""
            Review = review.get_text(strip=True) if review else ""
        
            data.append([Name, Title, Rating, City, Month, Review])
            
            #name=info.select_one('div.css-146c3p1') 
            #title= info.select_one('div.css-146c3p1')
            #rating= info.select_one('div.css-146c3p1')
            #city= info.select_one('div.css-146c3p1')
            #Date= info.select_one('div.css-146c3p1 r-dnmrzs r-1udh08x r-1udbk01 r-3s2u2q r-1iln25a')
            #review= info.select_one('div.G4PxIA')
            #Name=name.get_text()
            #Title=title.get_text()
            #Rating=rating.get_text()
            #City=city.get_text()
            #Month=Date.get_text()
            #Review=review.get_text()
            #data.append([Name,Title,Rating,City,Month,Review])
            data.append([Name, Title, Rating, City, Month, Review])
```

```python
df = df.iloc[2:].reset_index(drop=True)
df.head()
reviews=df['1']
print(reviews.head())
```

```python
import re

def clean_text(text):

    text = str(text).lower()

    text = re.sub(r'[^a-zA-Z ]', '', text)

    text = re.sub(r'\s+', ' ', text)

    return text.strip()

reviews = reviews.apply(clean_text)

print(reviews.head())
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()

tokenizer.fit_on_texts(reviews)
```
