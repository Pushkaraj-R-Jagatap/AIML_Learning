```python
import requests
from bs4 import BeautifulSoup
url=input("Enter the url")
page=1
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "en-IN,en;q=0.9",
    "Referer": "https://www.flipkart.com/"
}
while True:
    site="{}&page={}".format(url,page)
    res=requests.get(site,headers=headers)
    soup=BeautifulSoup(res.text,"lxml")
    print(res)
    review=soup.select('div.col.x_CUu6')


    if not review:
        print("no valid page")
        break
    with open(f'scraped_data/review{page}.html',"w",encoding='utf-8')as f:
        f.write(res.text)
        print(f"downloading the date from page no{page}")

    page=page+1

# from selenium import webdriver
# from selenium.webdriver.common.by import By
# from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager
# import time

# url = input("Enter Flipkart review URL: ")

# driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
# driver.get(url)
# time.sleep(3)

# page = 1

# while True:
#     print(f"Saving page {page}")

#     with open(f"scraped_data/review{page}.html", "w", encoding="utf-8") as f:
#         f.write(driver.page_source)

#     try:
#         next_btn = driver.find_element(By.XPATH, "//span[text()='Next']")
#         next_btn.click()
#         time.sleep(3)
#         page += 1
#     except:
#         print("No more pages")
#         break

# driver.quit()
```

```python
data=[]
for i in range (1,3):
    with open(f'scraped_data/review{i}.html','r',encoding='utf-8') as f:
        html_content=f.read()
        soup=BeautifulSoup(html_content,'html.parser')
        all_info=soup.select('div.col.x_CUu6')
       

        for info in all_info:
            
            name=info.select_one('p.zJ1ZGa.ZDi3w2') 
            title= info.select_one('p.qW2QI1')
            rating= info.select_one('div.MKiFS6.ojKpP6')
            city= info.select_one('p.Zhmv6U span:nth-of-type(2)')
            Date= info.select_one('div.row p.zJ1ZGa:last-of-type')
            review= info.select_one('div.G4PxIA')
            Name=name.get_text()
            Title=title.get_text()
            Rating=rating.get_text()
            City=city.get_text()
            Month=Date.get_text()
            Review=review.get_text()
            data.append([Name,Title,Rating,City,Month,Review])
            

data
```
