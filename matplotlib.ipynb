```python
oscar_movies=["The Dark Knight",
              "The Hurt Locker",
              "The King's Speech",
              "The Artist",
              "Argo"
             ]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] #in SM

plt.bar(years,oscar_revenue,color="green")
plt.title("Revenue in each Year for Oscar Movie")
plt.xlabel("year")
plt.ylabel("Revenue in ($M)")
plt.ylim(0,max(oscar_revenue)+100)

for i in range (len (years)):
    plt.text(years[i],oscar_revenue[i]+10,str(oscar_revenue[i]),ha="center")
```

```python
import matplotlib.pyplot as plt
import numpy as np
oscar_movies=["The Dark Knight",
              "The Hurt Locker",
              "The King's Speech",
              "The Artist",
              "Argo"
             ]
non_oscar_movies=["Avatar",'inception',"hugo","lincoln","salar"]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] #in SM
non_oscar_revenue=[500,343,453,567,354]
x=np.arange(len(years))
width=0.4

plt.bar(x-width/2,oscar_revenue,width,label="Oscar Movies")
plt.bar(x+width/2,non_oscar_revenue,width,label="Non-Oscar Movies")
plt.title("Oscar vs non-oscar movies")
plt.xlabel("Years")
plt.ylabel("Revnue in $M")
plt.legend()
plt.xticks(x,years)
```
