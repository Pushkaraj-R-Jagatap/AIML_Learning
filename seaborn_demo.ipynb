```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
sns.get_dataset_names()

tips=sns.load_dataset("tips")
# print(tips.head())
# print(tips.tail())

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    size='size',
    style='smoker')
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    kind='line',
    y='tip',
    hue='smoker',
    )
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time")
```

```python
#line plot
flight=sns.load_dataset("flights")
flight.head()
sns.lineplot(
    data=flight,
    x="year",
    y="passengers",
    hue="month")
```

```python
print(tips.head())
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex")
```

```python
#box plot
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
    )
```

```python
#histogram'
pehguins=sns.load_dataset("penguins")
print(pehguins.head())
sns.histplot(
    data=pehguins,
    x="body_mass_g",
    bins=30)
```

```python
#flights data
sns.get_dataset_names()
flights=sns.load_dataset("flights")
print(flights.head())
data=flights.pivot(index="month",columns='year',values='passengers')
sns.heatmap(
    data,
    cmap="coolwarm",
    annot=True,
    fmt="d")
plt.title("Passenger Heatmap")
plt.tight_layout()
     
```
