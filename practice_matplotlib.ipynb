```python
#task
#solution no1
import matplotlib.pyplot as plt
import numpy as np
days=["Mon","Tue","Wed","Thu","Fri"]
cities=["New york","london","delhi","tokyo"]
city_no=0
temperature=[
    [22,23,21,24,25],#new york
    [18,19,17,20,21],#london
    [30,32,31,33,34],#delhi
    [25,26,24,27,28]#tokyo
]

fig,axes=plt.subplots(2,2)
fig.tight_layout()
for i in range(2):
    for j in range(2):
        axes[i][j].plot(days,temperature[city_no])
        axes[i][j].set_title(cities[city_no])
        axes[i][j].grid(True)
        city_no=city_no+1
fig.suptitle("temperature in different cities")
fig.supylabel("Temperature")
fig.supxlabel("Days")
```

```python
#solution no 2
import matplotlib.pyplot as plt
import numpy as np
days=["Mon","Tue","Wed","Thu","Fri"]
cities=["New york","london","delhi","tokyo"]
city_no=0
temperature=[
    [22,23,21,24,25],#new york
    [18,19,17,20,21],#london
    [30,32,31,33,34],#delhi
    [25,26,24,27,28]#tokyo
]

fig,axes=plt.subplots(2,2)
axes=axes.flatten()
for i,ax in enumerate(axes):
    ax.plot(days,temperature[i],marker='o')
    ax.set_title(cities[i])
    ax.grid(True)

fig.suptitle("Temperature in cities over the week")
fig.supylabel("Temperature")
fig.supxlabel("Days")
fig.tight_layout()
```
