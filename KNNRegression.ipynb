```python
import numpy  as np 
class KNNRegressor:

    def __init__(self,k=2):
        self.k=k

    def fit(self,x,y):
        self.x_train=np.array(x)
        self.y_train=np.array(y)


    def _euclidean_distance(self,x1,x2):
        return np.sqrt(np.sum((x1-x2)**2))

    def _predict_one(self,x):
        #distence from all train ing points
        distances=[self._euclidean_distance(x,x_train)
                   for x_train in self.x_train]
        #indeices of k nearesst
        k_indices=np.argsort(distances)[:self.k]

        #average of their targets

        k_values =[self.y_train[i] for i in k_indices]
        return np.mean(k_values)

    def predict(self,X):
        return np.array([self._predict_one(x) for x in X])


x_train=np.array([[1],[2],[3],[4],[5]])
y_train=np.array([2,4,6,8,10])
x_test=np.array([[1.5],[3.5],[6]])
model=KNNRegressor(k=2)
model.fit(x_train,y_train)

predictions=model.predict(x_test)
print("predictions:",predictions)
```
