```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("DateFruit_Dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
 df.shape

```

```python
x=df.drop("Class",axis=1)
y=df["Class"]
```

```python
df['Class'].unique()
from sklearn.preprocessing import StandardScaler ,LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python


#ANN
import torch
import torch.nn as nn
import torch.optim as optim 
from torch.utils.data import DataLoader,TensorDataset
```

```python
x_train_tensor=torch.tensor(x_train_scaled,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train,dtype=torch.long)

x_test_tensor=torch.tensor(x_test_scaled,dtype=torch.float32)
y_test_tensor=torch.tensor(y_test,dtype=torch.long)
```

```python
train_dataset=TensorDataset(x_train_tensor,y_train_tensor)
test_dataset=TensorDataset(x_test_tensor,y_test_tensor)
```

```python
train_loder=DataLoader(train_dataset,batch_size=32,shuffle=True)
test_loader=DataLoader(test_dataset,batch_size=32)
```

```python
#Build teh model
class ANN(nn.Module):
    def __init__(self):
        super(ANN,self).__init__()
        self.model=nn.Sequential(
            nn.Linear(x.shape[1],64),
            nn.ReLU(),



            nn.Linear(64,64),
            nn.ReLU(),

            nn.Linear(64,7)
        )
    def forward(self,x):
        return self.model(x)
            
    
```

```python
model=ANN()
criteria=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters())
```

```python
#train the NN
epochs=100
for epoch in range (epochs):
    model.train()

    running_loss=0.0
    for xb,yb in train_loder:
        optimizer.zero_grad()
        outputs=model(xb)
        loss=criteria(outputs,yb)
        loss.backward()
        optimizer.step()#parameter uipdate
        running_loss +=loss.item()

    train_loss=running_loss/len(train_loder)
    print(f"epoch ={epoch}/{epochs},loss={train_loss}")
    
```

```python
#evaluate
model.eval()
total=0
correct=0
with torch.no_grad():
    for xb,yb in test_loader:
        outputs=model(xb)
        max_val,predicted=torch.max(outputs,1)
        correct+=(predicted==yb).sum().item()
        total+=yb.size(0)  #actual sample in each batch
print("total value:",total)
print("correct values:",correct)
print("accuracy :",correct/total*100)
```
