```python
import pandas as pd
df=pd.read_csv("raw_data.csv")
df2=df.copy()
df2["tax"]=df2["income"].apply(lambda x: "20%" if x>=60000 else "10%")
gender_map={"Male":"M","Female":"F","Unknown":"U"}
df2['gender']=df2['gender'].map(gender_map)
df2.assign(newe_salary=df2['income']*1.2)
df2['country']=df2['country'].replace("USA","United state")
df2.rename(columns={"income":"Salary"})
df2.rename(index={1:"First"})
df2=df2.fillna(50)
sorteed_valuees=df2.sort_values(['income',"age"])
sorteed_valuees.sort_index()

sorteed_valuees.reset_index(drop=True)
sorteed_valuees["Ranking"]=sorteed_valuees["income"].rank(ascending=False,method="dense")
sorteed_valuees[["id","name","age","country","gender","income","Ranking","tax"]]

df2=df.copy()
new_col_order=[col for col in df2.columns if col !="id"]+["id"]
print(new_col_order)
df2[new_col_order]

csv=pd.read_csv("IPL2025Batters.csv")
sort=csv.sort_values("HS",ascending=False)
sort
```

```python
df=pd.read_csv("raw_data.csv")
df2=df.copy()
df2=df2.drop_duplicates()
df2=df2.fillna(0)
df2=df2.sort_values("income")
df2=df2.reset_index(drop=True)
df2.to_csv("sorteddata.csv")
df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("gender")["income"].max()
df.groupby("country")["income"].agg(avg_salary="mean",max_salary="max",min_salary="min")
df.groupby("country").agg(
    {"income":"mean",
     "age":"mean",}
)
df.groupby("country").agg({
'income':'max',
'age':"max"
})

```

```python
df_customere=pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name":["adam","bob","charlie","dave"]
})

df_ordere=pd.DataFrame({
    "order_id":[101,102,103,104],
    "customer_id":[1,2,7,9],
    "amout":[123,676,67,467]
})

ineer_join=pd.merge(df_customere,df_ordere,on="customer_id")   #inner join
left_join=pd.merge(df_ordere,df_customere,on="customer_id",how="left")
right_join=pd.merge(df_customere,df_ordere,on="customer_id",how="right")
outer_join=pd.merge(df_ordere,df_customere,on="customer_id",how="outer")
print(ineer_join)
print(left_join)
print(right_join)
print(outer_join)
```
