```python
import matplotlib.pyplot as plt
import numpy as py
import seaborn as sns
fig,ax=plt.subplots()
tips=sns.load_dataset("tips")
sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    # errorbar=None,
    ax=ax)
plt.title("Avg Bill Amount")
plt.xlabel("Days")
plt.ylabel("Total Tip amount")
plt.tight_layout()
```
