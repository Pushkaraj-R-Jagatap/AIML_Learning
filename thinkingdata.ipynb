```python
import json
```

```python
 #load the data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
        
    return data
    
    
```

```python
data=load_data("store_data.json")
print(data)
print(type(data))
```

```python
#clean data
def clean_data(data):
    text_to_number={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    for user in data:
       #clean ratings
        raw_rating=str(user["rating"]).strip().lower()
        if(raw_rating in text_to_number):
            raw_rating=text_to_number[raw_rating]
        user["rating"]=raw_rating
       


        #handle missing value
        raw_age=user.get('age')
        if(raw_age==None):
            user["age"]=None

        #deduplication
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
        
```

```python
data=clean_data(data)
print(data)
```

```python
#get meaningful insights from data
def get_insights(data):
    #avg rating
    tot_rating=0
    for user in data:
        tot_rating +=float(user["rating"])
    print(f"average rating is {tot_rating/len(data)}")
    #percentages of users with poor rating
    poor_rating=0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating+=1
    print(f"% of user with poor rating={poor_rating/len(data)*100}%")
    
```

```python
get_insights(data)
```

```python
#to build the recommendation features
def get_recommendations(data):
    recommendation=[]
    print(type(recommendation))
    for user in data:
        curr_recomm={}
        curr_recomm["name"]=user["name"]

        
        if(float(user["rating"])>=4):
            curr_recomm["brand"]="Apple"
        else:
            curr_recomm["brand"]="Samsung"
        recommendation.append(curr_recomm)
    return recommendation
```

```python
get_recommendations(data)
```
