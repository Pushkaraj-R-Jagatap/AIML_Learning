```python
#Question 1
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data =np.random.normal(loc=70,scale=10,size=200)
plt.hist(data)
plt.axvline(x=70,label="Mean of the Data",color="red")
```

```python
#Question 2
data=sns.load_dataset("penguins")
sns.lineplot(
    data=data,
    x="flipper_length_mm",
    y="body_mass_g",
    hue='species'
    
   
)
plt.title("flipper_length_mmbody_mass_g")
```

```python
#Question 3
tips=sns.load_dataset("tips")
print(tips.head())
sns.boxplot(data=tips,
            x="day",
            y="total_bill",
            hue="sex")
```

```python
#question 4
print(tips.head())
sns.scatterplot(data=tips,
                x="total_bill",
                y="tip",
                hue="sex",
                style="smoker")
plt.title("total_billtip")
plt.ylabel("Tip Amount")
plt.xlabel("Total Amount")
```

```python
#question 5
data=tips.pivot_table(index="day",columns="time",values="size")
sns.heatmap(
    data,
    cmap="coolwarm",
    annot=True,
    )
plt.tight_layout()
```
