```python
import requests
from bs4 import BeautifulSoup
page=1
while True:
    url=f"https://quotes.toscrape.com/page/{page}/"
    res=requests.get(url)
    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select('div.quote')
 

    if not quotes:
        print("no valid page")
        break
    with open(f"scraped_data/page{page}.html","w",encoding="utf-8")as f:
        f.write(res.text)
        print(f"downloaded the data from page{page}")
    page=page+1
```

```python
# for i in range(1,11):
    # with open(f"scraped_data/page{i}.html","r",encoding="utf-8") as f:
    #      html_content=f.read()
    #      soup=BeautifulSoup(html_content)
all_html = []      # store raw HTML
# all_soups = []     # store parsed HTML
all_tags=[]
data=[]
for i in range(1, 11):
    with open(f"scraped_data/page{i}.html", "r", encoding="utf-8") as f:
        html_content = f.read()
        # all_html.append(html_content)
        soup=BeautifulSoup(html_content, "html.parser")

        all_quotes=soup.select("div.quote")

        for q in all_quotes:
            for tag in q.select(".tags .tag"):
                 all_tags.append(tag.get_text())
            if "life" in all_tags:
                text=q.select_one("span.text").get_text(strip=True)
                author=q.select_one("small.author").get_text(strip=True)
                data.append([text,author])
                all_tags=[]

```

```python
data
import pandas as pd
table=pd.DataFrame(data)
table.to_csv("scraped_data/table.csv",index=False)

df=pd.read_csv("scraped_data/table.csv")
df
```
