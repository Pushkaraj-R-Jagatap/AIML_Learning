```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
from torchvision.datasets import CIFAR10
```

```python
#dataset & dataloader
from torch.utils.data import DataLoader
import torchvision.transforms as transforms

#image => scale(0,1) =>normalize =>(-1,1)
transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])

trainset=CIFAR10(root="./Data",train=True ,download=True,transform=transform)
testnset=CIFAR10(root="./Data",train=False ,download=True,transform=transform)
```

```python
trainloader=DataLoader(trainset,batch_size=64,shuffle=True)
testloader=DataLoader(testnset,batch_size=64)
```

```python
#Build the CNN
class CNN(nn.Module):
    def __init__(self):
        super(CNN,self).__init__()

        self.conv_layers=nn.Sequential(
            nn.conv2d(3,32,kernel=3,padding=1)     ,    #3=input_channel,32=output_channel
            nn.ReLU(),
            nn.Maxpool2d(2,2),#kernel size=2,stride=2


            nn.conv2d(32,64,kernel=3,padding=1),        
            nn.ReLU(),
            nn.Maxpool2d(2,2), #kernel size=2,stride=2


            nn.conv2d(64,128,kernel=3,padding=1),         
            nn.ReLU(),
            nn.Maxpool2d(2,2), #kernel size=2,stride=2
        )

        self.fc_layer=nn.Sequential(
            nn.Linear(4*4*128,256),
            nn.ReLU(),
            nn.Linear(256,10),
        )
        def forward(self,x):
            x=self.conv_layers(x),
            x=x.view(x.size(0),-1),
            x=self.fc_layers(x)

            return x
```

```python
model=CNN()
criterion=nn.CrossEntropyLoss()
optimizer=Optim.Adam(model.parameters())
```
